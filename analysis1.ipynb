```python
import os
import pandas as pd

# Directory containing the CSV files
directory = 'datasets'

# List to store average rent for each locality
locality_avg_rent = []

# Loop through all CSV files in the directory
for file in os.listdir(directory):
    if file.endswith(".csv"):  # Check for CSV files
        file_path = os.path.join(directory, file)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path, encoding='ISO-8859-1')  # Adjust encoding if necessary
        
        # Check if 'rent' column exists and calculate the average rent for this locality
        if 'rent' in df.columns:
            avg_rent = df['rent'].mean()  # Calculate the average rent
            locality_name = os.path.splitext(file)[0]  # Remove the file extension to get the locality name
            locality_avg_rent.append((locality_name, avg_rent))  # Store locality name and average rent

# Convert the list to a DataFrame for better readability
avg_rent_df = pd.DataFrame(locality_avg_rent, columns=['Locality', 'Average Rent'])

# Find the locality with the highest average rent
max_rent_locality = avg_rent_df.loc[avg_rent_df['Average Rent'].idxmax()]

# Print the locality with the highest average rent
print(f"The locality with the highest average rent is: {max_rent_locality['Locality']} with an average rent of {max_rent_locality['Average Rent']}")
```
